```python
L = 10
T = 9

curant = 0.6

N_h = 100

h = L/N_h
tau = curant*h

N_tau = int(T/tau) + 1

a = 1

#Начальное
def zero_time_F(x):
    return np.sin(2*np.pi*x/L)
    #return 0;

#правое и левое
def zero_x_l_F(t):
    return zero_time_F(-a*t)
    #return 0;

def zero_x_r_F(t):
    return zero_time_F(L - a*t)
    #return 0;

# f
def f(x, t):
    return 0;

N_tau
```

## На посмотреть

### Схема Лакса

```python
#Генерируем нулевой слой
x_ans = [[zero_time_F(i*h) for i in range(0, N_h+1)]]
#print(x_ans)


for j in range(1, N_tau + 1):
    mid_res = [zero_x_l_F(tau*j)]
    for i in range (1 , N_h):
        mid_res = np.append(mid_res,  [tau*f(i*h, tau*j) - a*(x_ans[-1][i+1] - x_ans[-1][i-1])*(tau/(2*h)) + 0.5*(x_ans[-1][i+1] + x_ans[-1][i-1])])
    mid_res = np.append(mid_res, [zero_x_r_F(tau*j)])
    x_ans = np.append(x_ans, [mid_res], axis = 0)

#x_ans[-1]
```

```python



x1 = [np.linspace(0, L, num = N_h + 1)]
t1 = [np.linspace(0, T, num = N_tau + 1)]
x = x1
t = t1
for i in range(N_tau):
    x = np.append(x, x1, axis = 0)
for i in range(N_h):
    t = np.append(t, t1, axis = 0)
#print(x)
#print(np.transpose(t))

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('u')
ax.scatter(x, np.transpose(t), x_ans)
plt.show()
```

## Исследования

тут будут более оптимальные по памяти решения для ускорения, но они только невязку дадут

```python
L = 10
T = 9

curant = 0.3

N_h = 2000

h = L/N_h
tau = curant*h

N_tau = int(T/tau) + 1

a = 1

#Начальное
def zero_time_F(x):
    return np.sin(2*np.pi*x/L)
    #return 0;

#правое и левое
def zero_x_l_F(t):
    return zero_time_F(-a*t)
    #return 0;

def zero_x_r_F(t):
    return zero_time_F(L - a*t)
    #return 0;

def real_F(x, t):
    return zero_time_F(x - a*t)
    
# f
def f(x, t):
    return 0;

N_tau
```

### Схема Лакса

```python
#Генерируем нулевой слой
x_ans = [zero_time_F(i*h) for i in range(0, N_h+1)]
#print(x_ans)

delta = 0;

for j in range(1, N_tau + 1):
    mid_res = [zero_x_l_F(tau*j)]
    for i in range (1 , N_h):
        mid_res = np.append(mid_res,  [tau*f(i*h, tau*j) - a*(x_ans[i+1] - x_ans[i-1])*(tau/(2*h)) + 0.5*(x_ans[i+1] + x_ans[i-1])])
    mid_res = np.append(mid_res, [zero_x_r_F(tau*j)])
    real_ans = [real_F(i*h, j*tau) for i in range(0, N_h + 1)]
    delta = max(delta, np.sum((mid_res - real_ans)**2))
    x_ans = mid_res 

#x_ans[-1]
print(delta)
```
