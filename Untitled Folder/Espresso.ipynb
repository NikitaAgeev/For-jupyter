тут будут более оптимальные по памяти решения для ускорения, но они только невязку дадут

```python
L = 10
T = 9

curant = 1

N_h = 500

h = L/N_h
tau = curant*h

N_tau = int(T/tau) + 1

a = 1

#Начальное
def zero_time_F(x):
    return np.sin(2*np.pi*x/L)
    #return 0;

#правое и левое
def zero_x_l_F(t):
    return zero_time_F(-a*t)
    #return 0;

def zero_x_r_F(t):
    return zero_time_F(L - a*t)
    #return 0;

def real_F(x, t):
    return zero_time_F(x - a*t)
    
# f
def f(x, t):
    return 0;

N_tau
```

### Схема Лакса

```python
#Генерируем нулевой слой
x_ans = [zero_time_F(i*h) for i in range(0, N_h+1)]
#print(x_ans)

delta = 0;

for j in range(1, N_tau + 1):
    mid_res = [tau*f(0, tau*j) - a*(x_ans[1] - x_ans[-2])*(tau/(2*h)) + 0.5*(x_ans[1] + x_ans[-2])]
    for i in range (1 , N_h):
        mid_res = np.append(mid_res,  [tau*f(i*h, tau*j) - a*(x_ans[i+1] - x_ans[i-1])*(tau/(2*h)) + 0.5*(x_ans[i+1] + x_ans[i-1])])
    mid_res = np.append(mid_res, [mid_res[0]])
    real_ans = [real_F(i*h, j*tau) for i in range(0, N_h + 1)]
    delta = max(delta, np.max((mid_res - real_ans)**2))
    x_ans = mid_res 

#x_ans[-1]
print(np.sqrt(delta))
```

### Схема Лакса-Вендрофа

```python
#Генерируем нулевой слой
x_ans = [zero_time_F(i*h) for i in range(0, N_h+1)]
#print(x_ans)

delta = 0;

for j in range(1, N_tau + 1):
    mid_res = [tau*f(0, tau*j) - a*(x_ans[1] - x_ans[-2])*(tau/(2*h)) + x_ans[0] + (a**2 *tau**2/(2*h**2))*(x_ans[1] - 2*x_ans[0] + x_ans[-2])]
    for i in range (1 , N_h):
        mid_res = np.append(mid_res,  [tau*f(i*h, tau*j) - a*(x_ans[i+1] - x_ans[i-1])*(tau/(2*h)) + x_ans[i] + (a**2 *tau**2/(2*h**2))*(x_ans[i+1] - 2*x_ans[i] + x_ans[i-1])])
    mid_res = np.append(mid_res, [mid_res[0]])
    real_ans = [real_F(i*h, j*tau) for i in range(0, N_h + 1)]
    delta = max(delta, np.max((mid_res - real_ans)**2))
    x_ans = mid_res 

#x_ans[-1]
print(np.sqrt(delta))
```
