```python
def f(x, y):
    return -32*(x-x*x + y-y*y)

eps = 10**-6
N_x = 30
N_y = N_x

h_x = 1/N_x
h_y = 1/N_y

l = 2*np.pi*np.pi
L = 8*N_x*N_x

tau = 2/(l + L)
q = (L - l)/(L + l)

print("tau:",tau)
print("q:",q)

def f_solve(x, y):
    return (x**4 - 2*x**3 + y**4 - 2*y**3)*(8/3)
```

```python
x1 = np.array([np.linspace(0, 1, num = N_x + 1)])
y1 = np.array([np.linspace(0, 1, num = N_y + 1)])
x = x1
y = y1
for i in range(N_x):
    x = np.append(x, x1, axis = 0)
for i in range(N_y):
    y = np.append(y, y1, axis = 0)

y = np.transpose(y)
    
u = np.array([np.array([0.0 for i in range (0, N_x + 1)]) for j in range(0, N_y + 1)])

#print(x[2:0])

delta = 1

cnt = 0

while delta > ((1-q)/q)*eps:
    u_next = np.zeros((N_x + 1, N_y + 1), dtype = float )
    
    u_next[1:-1, 1:-1] = u[1:-1, 1:-1] + tau*((u[2:, 1:-1] - 2*u[1:-1, 1:-1] + u[0:-2, 1:-1])/(2*h_y) + (u[1:-1, 2:] - 2*u[1:-1, 1:-1] + u[1:-1, 0:-2])/(2*h_x) - f(x[1:-1, 1:-1], y[1:-1, 1:-1]))
    #print(u_next)
    delta = np.sqrt(np.max((np.array(u_next) - np.array(u))**2))
    u = u_next
    cnt = cnt + 1
    if (cnt == 10000):
        print("Грустно")
    if (cnt == 100000):
        print("Очень Грустно")
print(cnt)

#f(x[1:-1, 1:-1], y[1:-1, 1:-1])
#u_next
```
