```python
def f(x, y):
    return -32*(x*(1-x) + y*(1-y))

eps = 10**-6
N_x = 20
N_y = N_x

h_x = 1/N_x
h_y = 1/N_y

l = 2*np.pi*np.pi
L = 8*N_x*N_x

tau = 2/(l + L)
q = (L - l)/(L + l)

print("tau:",tau)
print("q:",q)
```

```python
u = [[0 for i in range (0, N_x + 1)] for j in range(0, N_y + 1)]

delta = 1

#print(u)

cnt = 0

while delta > ((1-q)/q)*eps:
    u_next = [[0 for i in range (0, N_x + 1)] for j in range(0, N_y + 1)]
    for j in range(1, N_y):
        for i in range(1, N_x):
            #print(j, i)
            #print(u_next)
            u_next[j][i] = u[j][i] + tau*((u[j+1][i] - 2*u[j][i] + u[j-1][i])/(2*h_y) + (u[j][i+1] - 2*u[j][i] + u[j][i-1])/(2*h_x) - f(i*h_x, j*h_y))
    delta = np.sqrt(np.max((np.array(u_next) - np.array(u))**2))
    u = u_next
    cnt = cnt + 1
    if (cnt == 10000):
        print("Грустно")
    if (cnt == 100000):
        print("Очень Грустно")
print(cnt)

#x
```

```python
x1 = [np.linspace(0, 1, num = N_x + 1)]
y1 = [np.linspace(0, 1, num = N_y + 1)]
x = x1
y = y1
for i in range(N_x):
    x = np.append(x, x1, axis = 0)
for i in range(N_y):
    y = np.append(y, y1, axis = 0)
#print(x)
#print(np.transpose(t))

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('u')
ax.scatter(x, np.transpose(y), u)
plt.show()
```
